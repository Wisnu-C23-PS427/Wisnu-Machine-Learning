```python
import re

# Fungsi untuk membersihkan gelar
def clean_title(name):
    pattern = r'([A-Za-z\s]+)(?:,\s[A-Za-z\.\s]+)?$'  # Pola regex untuk mencocokkan dan mengambil hanya nama
    match = re.search(pattern, name)
    cleaned_name = match.group(1) if match else ""  # Mengembalikan string kosong jika tidak ada kecocokan
    return cleaned_name

# Membersihkan kolom Nama
df_local_guide['Nama Pemandu'] = df_local_guide['Nama Pemandu'].apply(clean_title)

# Menampilkan dataframe hasil
df_local_guide.head()
```

create preference

```python
df_local_guide['Preferensi'] = df_local_guide[['Lokasi', 'Category', 'Pengalaman Objek Wisata']].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)
```

```python
df_local_guide = df_local_guide[['Nama Pemandu', 'Rating', 'Preferensi']]
df_local_guide
```

```python
# keyword
list_keywords = ['Budaya', 'Taman Hiburan', 'Cagar Alam', 'Bahari', 'Pusat Perbelanjaan', 'Tempat Ibadah'] 

# Fungsi untuk mencari kata dalam preferensi
def cari_kata(preferensi):
    kata_ditemukan = []
    preferensi_lower = preferensi.lower() # Mengubah preferensi menjadi huruf kecil
    
    for keywords in list_keywords:
        keywords_lower = keywords.lower() # Mengubah keywords menjadi huruf kecil
        
        if keywords_lower in preferensi_lower:
            kata_ditemukan.append(keywords)
    
    return kata_ditemukan

# Membuat kolom keywords
df_local_guide['keywords'] = df_local_guide['Preferensi'].apply(cari_kata)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Mengambil nilai rating dari dataframe
ratings = df_local_guide['Rating'].values.reshape(-1, 1)

# Melakukan normalisasi fitur rating
normalized_ratings = scaler.fit_transform(ratings)

# Menyimpan hasil normalisasi ke dalam kolom "normalized_rating"
df_local_guide['Rating'] = normalized_ratings

# Tampilkan dataframe dengan fitur rating yang telah dinormalisasi
df_local_guide
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

class Tourist:
    def __init__(self, name, needs):
        self.name = name
        self.needs = needs

class Guide:
    def __init__(self, name, expertise, rating):
        self.name = name
        self.expertise = expertise
        self.rating = rating

class Matchmaking:
    def __init__(self):
        self.tourists = []
        self.guides = []
        self.vectorizer = TfidfVectorizer()

    def add_tourist(self, tourist):
        self.tourists.append(tourist)

    def add_guide(self, guide):
        self.guides.append(guide)

    def preprocess_text(self, text):
        stop_words = set(stopwords.words('english') + stopwords.words('indonesian'))
        tokens = word_tokenize(text.lower())
        filtered_tokens = [token for token in tokens if token.isalpha() and token not in stop_words]
        preprocessed_text = ' '.join(filtered_tokens)
        return preprocessed_text

    def fit_vectorizer(self):
        corpus = []
        for tourist in self.tourists:
            corpus.append(self.preprocess_text(tourist.needs))
        for guide in self.guides:
            corpus.append(self.preprocess_text(guide.expertise))
        self.vectorizer.fit(corpus)

    def find_guide(self, tourist):
        tourist_needs = self.preprocess_text(tourist.needs)
        tourist_vector = self.vectorizer.transform([tourist_needs])

        matched_guides = []
        for guide in self.guides:
            guide_expertise = self.preprocess_text(guide.expertise)
            guide_vector = self.vectorizer.transform([guide_expertise])

            similarity_score = cosine_similarity(tourist_vector, guide_vector)[0][0]
            if similarity_score > 0:
                matched_guides.append((guide, similarity_score))
        matched_guides = sorted(matched_guides, key=lambda x: (x[1] + x[0].rating), reverse=True)

        return matched_guides
```

```python
import warnings

# Ignore future warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Initialize
platform = Matchmaking()

# Menambahkan tourists dan guides ke platform
tourist1 = Tourist("tourist 1", "Bagaimana cara saya bisa mengunjungi Wisata Budaya?")
tourist2 = Tourist("tourist 2", "Saya membutuhkan bimbingan dalam melakukan perjalanan ke tempat budaya")
tourist3 = Tourist("tourist 3", " Saya ingin melihat kawasan cagar alam di Indonesia")
platform.add_tourist(tourist1)
platform.add_tourist(tourist2)
platform.add_tourist(tourist3)

for index, row in df_local_guide.iterrows():
    guide_expertise = ' '.join(row['Preferensi'].split()[:10])  # Mengambil maksimal 10 kata dari guide.expertise
    guide = Guide(row['Nama Pemandu'], guide_expertise, row['Rating'])
    platform.add_guide(guide)

# Fitting vectorizer
platform.fit_vectorizer()

# Melakukan pencocokan tourist dengan guide
for tourist in platform.tourists:
    matched_guides = platform.find_guide(tourist)
    if len(matched_guides) > 0:
        print(f"tourist '{tourist.name}' cocok dengan guide:")
        for guide, similarity_score in matched_guides:
            print(f"- {guide.name} ({guide.expertise}) - Similarity Score: {similarity_score} - Rating: {guide.rating}")
    else:
        print(f"tourist '{tourist.name}' tidak ditemukan guide yang cocok.")
```

```python
# Change to DataFrame

# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['Tourist', 'Guide', 'Expertise', 'Similarity Score', 'Rating'])

# Melakukan pencocokan tourist dengan guide
for tourist in platform.tourists:
    matched_guides = platform.find_guide(tourist)
    if len(matched_guides) > 0:
        for guide, similarity_score in matched_guides:
            # Append the result to the DataFrame
            results_df = results_df.append({
                'Tourist': tourist.name,
                'Guide': guide.name,
                'Expertise': guide.expertise,
                'Similarity Score': similarity_score,
                'Rating': guide.rating
            }, ignore_index=True)
    else:
        # Append the result to the DataFrame
        results_df = results_df.append({
            'Tourist': tourist.name,
            'Guide': 'N/A',
            'Expertise': 'N/A',
            'Similarity Score': 'N/A',
            'Rating': 'N/A'
        }, ignore_index=True)

# Print the DataFrame
results_df
```
